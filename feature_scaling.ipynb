# Normalization (Min - Max Normalization)

        normalization = x - xmin/(xmax-xmin)

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
scale.fit_transform(dataset[['SepalLengthCm','SepalWidthCm','PetalLengthCm']])
```

# Standardization(z-score Normalization)
       mean = 0 and standard deviation is 1

    std(z) = (x - mean) / standard_deviation 

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdscale = StandardScaler()
```

```python
stdscale.fit_transform(dataset[['SepalLengthCm','SepalWidthCm','PetalLengthCm']])
```
